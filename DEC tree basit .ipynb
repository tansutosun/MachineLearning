```python
#Import Organics csv file
import pandas as pd
inputdf=pd.read_csv("C:/IDS/TRAINING/Applied Data Science Techniques with Python/Course_Data/organics.csv")
inputdf=inputdf.dropna()
#read_excel
#
```

```python
X_Input_Num=inputdf[['DemAge','PromTime']]
Y_Target=inputdf[['TargetBuy']]
```

```python
#Now LEts run a decision tree Algorithm
#Load required libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn import tree
#Train the model with using Decision Tree Classifier

Clf=DecisionTreeClassifier(criterion='gini',max_depth=2)
clf1=Clf.fit(X_Input_Num,Y_Target)
tree.plot_tree(clf1) 


feat_importance = pd.DataFrame(clf1.tree_.compute_feature_importances(normalize=False))
degiskenler=pd.DataFrame(X_Input_Num.columns.values)
degiskenler
feat_importance
mergedDf = degiskenler.merge(feat_importance, left_index=True, right_index=True)
print(mergedDf)

tree.plot_tree(clf1, fontsize=8)
text_representation = tree.export_text(clf1)
```
