```python
# HO : MEAN=15
# Ha : MEAN <> 15 
# Test if 1 box is equal to 15 onces 
import scipy.stats as stats

stats.ttest_1samp(cereals.weight,popmean=15.00)
# H0: kutu=15 olma olasılığı  0,00000959 ==> % 0.000959
# Ha : kutu <> 15  ....>> 1- 0,00000959 : 0.999999 == > %99.99941
```

```python

# H0: Her ikigrup eşittir
# Ha: İki grup birbiri ile eşit değildir. 

# ANOVA  Morning markası ile RisenShien farklımıdır? 
Morning=cereals[cereals['brand']=='Morning']
RisenShine=cereals[cereals['brand']=='Rise n Shine']

import scipy.stats as stats
stat, p= stats.f_oneway(Morning['weight'],RisenShine['weight'])

# Interpret

alpha=0.05

if p>alpha:
    print('Two samples are identical (fail to reject H0)')
else: 
    print('Two samples are different each other (reject H0)')
```
