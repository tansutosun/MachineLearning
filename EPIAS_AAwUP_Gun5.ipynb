```python
organics
# Target Buy = A + B1*DemAge + B2*PromSpend + B3*DemGender + B4*DemCluster Group
# f(DemClusterGroup)=mean(Target)
organics.count()
organics.groupby(['DemGender']).count().TargetBuy
X=organics.groupby(['DemGender']).mean().TargetBuy
```

```python
X
```

```python
type(X)
```

```python
X=X.to_frame()
```

```python
X
X=X.rename(columns={'TargetBuy': 'Gender_FCT'})
X
```

```python
# How to Combine Data Frames 
organics
```

```python
inputdf=pd.merge(organics, X, how='left', on='DemGender')
```

```python
inputdf
```

```python
# DemClusterGroup
DCG=inputdf.groupby(['DemClusterGroup']).mean().TargetBuy
```

```python
DCG=DCG.to_frame()
DCG.rename(columns={'TargetBuy':'DCG_FCT'}, inplace=True)
```

```python
inputdf=pd.merge(left=inputdf, right=DCG, how='left', on='DemClusterGroup')
```
